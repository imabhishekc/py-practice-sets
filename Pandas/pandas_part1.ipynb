# Pandas

In computer programming, pandas is a software library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series.

## Pandas main Data-Structures
### 1. Dataframes
In pandas, suppose we take an excel file or data file that has multiple columns and rows that store values and data. Panadas call that entire table as Dataframe.
### 2. Series
The particular selected, row or a column in a dataset is called series in pandas.

```python
# Masking
mask = data['city'] == 'Hyderabad'
data[mask].shape[0]
```

```python
def get_city_match_count(city):
    mask = data['city'] == city
    return data[mask].shape[0]
```

```python
get_city_match_count('Rajkot')
```

```python
mask1 = data['city'] == 'Hyderabad'
mask2 = data['date'] > '2010-01-01'

data[mask1 & mask2]
data[mask1 & mask2].shape[0]
```
