```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Draw a 2D 'flat' space
ax.plot([1, 9], [5, 5], color='black', linewidth=2, label='Flat Spacetime')

# Draw two distant points (A and B)
ax.plot(2, 5, 'bo', markersize=10)  # Point A
ax.text(1.9, 5.3, 'A', fontsize=12)

ax.plot(8, 5, 'ro', markersize=10)  # Point B
ax.text(7.9, 5.3, 'B', fontsize=12)

# Draw a curved arc to represent a wormhole path (shortcut)
wormhole = patches.Arc((5, 5), 6, 4, angle=0, theta1=0, theta2=180, linestyle='solid', color='green', linewidth=2)
ax.add_patch(wormhole)
ax.text(4.9, 7.1, 'Wormhole', fontsize=12, color='green')

# Optional: draw a dashed line along the flat space to represent traditional path
ax.plot([2, 8], [5, 5], 'k--', label='Traditional Path')

# Aesthetic settings
ax.set_xlim(0, 10)
ax.set_ylim(3, 8)
ax.axis('off')
ax.legend()
plt.title("2D Wormhole Analogy – Folded Spacetime", fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create wormhole geometry
def wormhole(r_min=1, r_max=5, steps=100):
    u = np.linspace(-5, 5, steps)
    v = np.linspace(0, 2 * np.pi, steps)
    u, v = np.meshgrid(u, v)

    # Throat shape
    r = np.sqrt(u**2 + r_min**2)

    x = r * np.cos(v)
    y = r * np.sin(v)
    z = u
    return x, y, z

x, y, z = wormhole()
```

```python
# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='plasma', edgecolor='k', alpha=0.8)

ax.set_title("3D Wormhole Structure (Einstein-Rosen Bridge)")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```
