```python
# 2D Wormhole Analogy
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Flat space (2D representation of spacetime)
ax.plot([1, 9], [5, 5], 'k-', lw=2, label='Flat Spacetime')

# Two points in space (A and B)
ax.plot(2, 5, 'bo', markersize=10)
ax.text(1.8, 5.3, 'A', fontsize=12)

ax.plot(8, 5, 'ro', markersize=10)
ax.text(7.8, 5.3, 'B', fontsize=12)

# Wormhole shortcut arc
wormhole = patches.Arc((5, 5), 6, 4, angle=0, theta1=0, theta2=180, color='green', lw=2)
ax.add_patch(wormhole)
ax.text(4.7, 7.2, 'Wormhole Path', fontsize=12, color='green')

# Traditional path (flat)
ax.plot([2, 8], [5, 5], 'k--', lw=1.5, label='Traditional Path')

ax.set_xlim(0, 10)
ax.set_ylim(3, 8)
ax.axis('off')
ax.legend()
plt.title("2D Analogy: Wormhole as a Fold in Space", fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def wormhole_geometry(r_min=1, steps=150):
    u = np.linspace(-5, 5, steps)
    v = np.linspace(0, 2 * np.pi, steps)
    u, v = np.meshgrid(u, v)

    # Wormhole throat using Morris-Thorne metric simplification
    r = np.sqrt(u**2 + r_min**2)
    x = r * np.cos(v)
    y = r * np.sin(v)
    z = u
    return x, y, z
```

```python
x, y, z = wormhole_geometry()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='plasma', edgecolor='none', alpha=0.95)

ax.set_title("3D Wormhole Structure (Einstein-Rosen Bridge)", fontsize=14)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
